# Third function

```python
start_date = input("Enter start date range in YYYY-MM-DD Format: ")    #'2013-02-08'
end_date = input("Enter end date range in YYYY-MM-DD Format: ")      #'2013-02-14'
company_name = input("Company name: ")
data[(data['Name']==company_name) & (data['date']>=start_date) & (data['date']<=end_date)]['close'].max()
```

```python
def third(start_date,end_date,company_name):
    return data[(data['Name']==company_name) & (data['date']>=start_date) & (data['date']<=end_date)]['close'].max()
```

```python
third('2013-02-08','2013-02-14','ZTS')
```

# find maximum volatile value on a given date

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
date = input("Enter date for finding most volatile stock: ")
temp = data[data['date']==date]
temp['volatile'] = temp['open']-temp['close']
temp['volatile'] = temp['volatile'].apply(lambda x:abs(x))
temp[temp['volatile']==temp['volatile'].max()]
```
